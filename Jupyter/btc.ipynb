```python
import pandas as pd
import requests
#!pip install yfinance
import yfinance as yf
```

```python
# Loading Fear and Greed Index data via API and converting to a DataFrame
url = requests.get('https://api.alternative.me/fng/?limit=0')
data = url.json()['data']
df = pd.DataFrame(data)
df.head()
```

```python
# Checking data types
df.info()
```

```python
# Changing data types
df['timestamp'] = pd.to_numeric(df['timestamp'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit = 's')
df['value'] = df['value'].astype(int)
```

```python
df.tail(5)
```

```python
# Loading BTC price data
btc_finance = yf.download ('BTC-USD', start = '2018-02-05')
btc_finance.head()
```

```python
# Formatting data to extract columns for analysis
btc = btc_finance.reset_index()
```

```python
btc.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
```

```python
btc.head()
```

```python
# Checking data types
btc.info()
```

```python
df.info()
```

```python
# Merging data into a single DataFrame
df_final = pd.merge(df, btc, left_on = 'timestamp', right_on = 'Date')
df_final.head()
```

```python
# Removing unnecessary columns
df_final = df_final.drop(columns = ['time_until_update', 'Date'])
```

```python
df_final.columns = df_final.columns.str.lower()
df_final.head()
```

```python
# Calculating summary statistics for overview
analysis = df_final.groupby('value_classification')['close'].mean().sort_values()
analysis
```
